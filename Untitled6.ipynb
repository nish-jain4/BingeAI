```python
df_2= df_2.merge(df_1, left_on='title',right_on='title')
df_2.head()
```

```python
df_2 = df_2[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
import ast


def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df_2['genres'] = df_2['genres'].apply(convert)
```

```python
df_2['genres']
```

```python
df_2['keywords'] = df_2['keywords'].apply(convert)
df_2['keywords']
```

```python
df_2['cast'][0]
```

```python
df_2['cast'] = df_2['cast'].apply(lambda x:[i['name'] for i in ast.literal_eval(x)][:3]) # only top 3 actors
```

```python
df_2['crew'] = df_2['crew'].apply(lambda x:[i['name'] for i in ast.literal_eval(x) if i['job'] == 'Director']) # only top 3 actors
```

```python
df_2['tags'] = df_2['genres'] + df_2['keywords'] + df_2['cast'] + df_2['crew']
```

```python
df_2['tags']
```

```python
df_2= df_2[['movie_id', 'title', 'overview', 'tags']]
```

```python
df_2['tags'] = df_2['tags'].apply(lambda x: " ".join(x))
```

```python
df_2['tags'] = df_2['tags'].apply(lambda x: x.lower())
```

```python
df_2.head()
```

```python

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=5000, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_2['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = df_2[df_2['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]
    movie_indices = [i[0] for i in sim_scores]
    return df_2['title'].iloc[movie_indices]
```

The `cosine_sim_df` DataFrame now contains the cosine similarity scores between each pair of movies. A higher score indicates greater similarity. Next, I will define a function to get movie recommendations based on these scores.
